```python
d_4 = pd.read_csv('/datadrive/Data_4_meta.tsv', sep='\t', low_memory=False)
```

```python
d_4["Taxon_ID"][0]
```

```python
a = pd.read_csv('/datadrive/taxidlineage.dmp', sep='|', low_memory=False, header=None)
```

```python
for taxid in d_4["Taxon_ID"]:
    lineage = a[a[0] == int(taxid)][1]
    try:
        if '10239' != lineage.values[0].split()[0]:
            print(taxid, lineage)
            break
    except:
        print(taxid, lineage)
```
